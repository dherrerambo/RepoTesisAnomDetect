```python
from datatools.snowflake.Querys import run_sql_query
import pandas as pd
import pyperclip
```

```python
database = "VP_INFORMACION"
schema   = "OUTLIERS"
```

```python
archivos = ['AF', 'US', 'AC', 'AP', 'AU', 'AH', 'AN', 'AM', 'AT']
archivos
```

```python
sql = f"show tables in {database}.{schema}"
tbs = run_sql_query(sql)
tbs = tbs[tbs["name"].str.endswith("ANON").fillna(False)==True]
tbs = tbs.loc[:,:"owner"].reset_index(drop=True)
tbs
```

```python
resp = dict()
```

```python
for tb in tbs["name"]:
    if tb=="RIPS_AF__ANON":
        pass
    else:
        sql = f"""alter table {database}.{schema}.{tb} drop column "id_carga + id_persona" """
        resp_ = dict(run_sql_query(sql).iloc[0])
        try:
            resp[tb].update({"borrar campo": resp_})
        except:
            resp.update({tb: {"borrar campo": resp_}})
```

```python
for tb in tbs["name"]:
    if tb=="RIPS_AF__ANON":
        pass
    else:
        sql = f"""alter table {database}.{schema}.{tb} add column "id_carga + id_persona" varchar(1000)"""
        resp_ = dict(run_sql_query(sql).iloc[0])
        try:
            resp[tb].update({"crear campo": resp_})
        except:
            resp.update({tb: {"crear campo": resp_}})
```

```python
for tb in tbs["name"]:
    if tb=="RIPS_AF__ANON":
        pass
    else:
        sql = f"""UPDATE {database}.{schema}.{tb} SET "id_carga + id_persona" = concat_ws('_', cast(id_carga as varchar), id_persona)"""
        try:
            pyperclip.copy(sql)
        except:
            pass
        resp_ = dict(run_sql_query(sql).iloc[0])
        try:
            resp[tb].update({"update campo": resp_})
        except:
            resp.update({tb: {"update campo": resp_}})
```
